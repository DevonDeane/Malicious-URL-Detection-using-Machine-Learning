```python
import pandas as p
import math
import re
from multiprocessing import Process

def query_len(url):
    findQS = [re.findall(r'\?\S+', url)]
    Str = ""
    for lis in findQS:
        for val in lis:
            for ch in val:
                Str += ch
    qsLen = len(Str)
    return qsLen

def spec_char_counter(url):
    dot = "."
    hyp = "-"
    und = "_"
    eq = "="
    fr = "/"
    qm = "?"

    dots, hyps, unds, eqs, frs, qms = 0,0,0,0,0,0

    for ch in url:
        if ch == dot:
            dots+=1
        elif ch == hyp:
            hyps+=1
        elif ch == und:
            unds+=1
        elif ch == eq:
            eqs +=1
        elif ch == fr:
            frs +=1
        elif ch == qm:
            qms +=1
    return [dots, hyps, unds, eqs, frs, qms]

def digit_counter(string):
    dgt=0
    letters=0
    for s in string:
        for c in s:
            if str(c).isdigit():
                dgt+=1
    return (dgt)

def entropy(string):
    # get probability of chars in string
    prob = [ float(string.count(c)) / len(string) for c in dict.fromkeys(list(string)) ]
    # calculate the entropy
    entropy = - sum([ p * math.log(p) / math.log(2.0) for p in prob ])
    return entropy

def create_dataset(data, n):
    new_data = p.DataFrame()
    count = 0
    for index, row in data.iterrows():
        url = row['url']
        feature_list = []
        feature_list.append(url)
    
        feature_list.append(len(url))
        if(re.match(r'^\S+(?:[0-9]{1,3}\.){3}[0-9]{1,3}\S+$', url)):
            ip_pres = True
        else:
            ip_pres = False
        feature_list.append(ip_pres)
        feature_list.append(query_len(url))
        feature_list.append(len(re.findall('(?=token=)', url)))
        feature_list.extend(spec_char_counter(url))
    
        keyword_list = ["secure", "account", "webscr", "login", "ebayisapi", "signin", "banking", 
                "confirm", "blog", "logon", "signon", "login.asp", "login.php", "login.htm", 
                ".exe", ".zip", ".rar", ".jpg", ".gif", "viewer.php", "link=", "getImage.asp", 
                "plugins", "paypal", "order", "dbsys.php", "config.bin", "download.php", ".js", 
                "payment", "files", "css", "shopping", "mail.php", ".jar", ".swf", ".cgi", ".php", 
                "abuse", "admin", ".bin", "www", "personal", "update", "verification"]

        for keyword in keyword_list:
            feature_list.append(keyword in url)

        feature_list.append(len(re.findall('(?=;)', url)))
        feature_list.append(len(re.findall('(\()', url)))
        feature_list.append(len(re.findall('(\))', url)))
        feature_list.append(len(re.findall('(?=%)', url)))
        feature_list.append(len(re.findall('(?=&)', url)))
        feature_list.append(len(re.findall('(?=@)', url)))
        feature_list.append(digit_counter(url))
        feature_list.append(entropy(url))
        feature_list.append(row['label'])
    
        features = p.DataFrame(feature_list)
        count += 1
        print(count)
        new_data = new_data.append(features.T, ignore_index=True)
    
    column_names = ["URL", "URL Length", "IP_Presence", "Query Length", "Number of Tokens", "Number of .", 
                "Number of -", "Number of _", "Number of =", "Number of /", "Number of ?", "secure", 
                "account", "webscr", "login", "ebayisapi", "signin", "banking", 
                "confirm", "blog", "logon", "signon", "login.asp", "login.php", "login.htm", 
                ".exe", ".zip", ".rar", ".jpg", ".gif", "viewer.php", "link=", "getImage.asp", 
                "plugins", "paypal", "order", "dbsys.php", "config.bin", "download.php", ".js", 
                "payment", "files", "css", "shopping", "mail.php", ".jar", ".swf", ".cgi", ".php", 
                "abuse", "admin", ".bin", "www", "personal", "update", "verification", "Number of ;", 
                "Number of Open Paranthesis", "Number of Close Paranthesis", "Number of %", "Number of &", 
                "Number of @", "Number of Digits", "Entropy", "Label"]
    new_data.columns = column_names
    new_data = new_data.replace(False, 0)
    new_data = new_data.replace(True, 1)
    new_data.to_csv("Extracted_Data"+n+".csv", encoding='utf-8', index=False)
    
def Main():
    
    data1 = p.read_csv(r'data.csv', skiprows=[i for i in range(105116,420465)])
    data2 = p.read_csv(r'data.csv', skiprows=[i for j in (range(1,105116), range(210233,420465)) for i in j])
    data3 = p.read_csv(r'data.csv', skiprows=[i for j in (range(1,210233), range(315349,420465)) for i in j])
    data4 = p.read_csv(r'data.csv', skiprows=[i for i in range(1,315349)])
    
    p1 = Process(target=create_dataset, args=(data1,"1",))
    p2 = Process(target=create_dataset, args=(data2,"2",))
    p3 = Process(target=create_dataset, args=(data3,"3",))
    p4 = Process(target=create_dataset, args=(data4,"4",))
    
    p1.start()
    p2.start()
    p3.start()
    p4.start()
    
    p1.join()
    p2.join()
    p3.join()
    p4.join()
    
    final_data = p.read_csv(r'Extracted_Data1.csv')
    final_data = final_data.append(p.read_csv(r'Extracted_Data2.csv'), ignore_index=True)
    final_data = final_data.append(p.read_csv(r'Extracted_Data3.csv'), ignore_index=True)
    final_data = final_data.append(p.read_csv(r'Extracted_Data4.csv'), ignore_index=True)
    final_data.to_csv("Extracted_Data.csv", encoding='utf-8', index=False)

if __name__ == '__main__':
    Main()
```
