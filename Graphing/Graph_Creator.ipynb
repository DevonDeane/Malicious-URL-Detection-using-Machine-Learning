```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt


PV_NFSE = pd.read_csv (r'Preferential_Voting_Using_NFSE.csv')
RF_ROS = pd.read_csv (r'Random_Forest_Using_RandomOversampler.csv')
RF_IMB = pd.read_csv (r'Random_Forest_Imbalanced.csv')
DT_ROS = pd.read_csv (r'Decision_Tree_Using_RandomOversampler.csv')
RF_SMOTE = pd.read_csv (r'Random_Forest_Using_SMOTE.csv')
```

```python
fig, ax = plt.subplots(figsize =[10, 5])
fig.suptitle("Malicious Class Classification Report",size=20)
# fig = plt.figure(figsize =[10, 5])
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color='blue', lw=4),
                Line2D([0], [0], color='red', lw=4),
                Line2D([0], [0], color='green', lw=4)]

meanpointprops = dict(marker='D', markeredgecolor='black',
                      markerfacecolor='gray')
ax.legend(custom_lines, ['F1-score', 'Recall', 'Precision'], loc='center left', bbox_to_anchor=(1, .9))


box_plot_data=[PV_NFSE['Malicious_Precision'],RF_ROS['Malicious_Precision'],RF_SMOTE['Malicious_Precision'],DT_ROS['Malicious_Precision'],RF_IMB['Malicious_Precision']]
box= ax.boxplot(box_plot_data,vert=0,patch_artist=True,labels=['PV-NFSE','RF-ROS','RF-SMOTE','DT-ROS','RF_IMB'],meanprops=meanpointprops,showmeans=True,widths=.4)


for patch in box['boxes']:
    patch.set_edgecolor('green')
    patch.set_facecolor('none')

box_plot_data=[PV_NFSE['Malicious_Recall'],RF_ROS['Malicious_Recall'],RF_SMOTE['Malicious_Recall'],DT_ROS['Malicious_Recall'],RF_IMB['Malicious_Recall']]
box= ax.boxplot(box_plot_data,vert=0,patch_artist=True,labels=['PV-NFSE','RF-ROS','RF-SMOTE','DT-ROS','RF_IMB'],meanprops=meanpointprops,showmeans=True,widths=.4)

for patch in box['boxes']:
    patch.set_edgecolor('red')
    patch.set_facecolor('none')
    
    
box_plot_data=[PV_NFSE['Malicious_F1-score'],RF_ROS['Malicious_F1-score'],RF_SMOTE['Malicious_F1-score'],DT_ROS['Malicious_F1-score'],RF_IMB['Malicious_F1-score']]
box= ax.boxplot(box_plot_data,vert=0,patch_artist=True,labels=['PV-NFSE','RF-ROS','RF-SMOTE','DT-ROS','RF_IMB'],meanprops=meanpointprops,showmeans=True,widths=.4)

for patch in box['boxes']:
    patch.set_edgecolor('blue')
    patch.set_facecolor('none')
fig.savefig("Benign_Malicious.jpeg", bbox_inches='tight')
```
